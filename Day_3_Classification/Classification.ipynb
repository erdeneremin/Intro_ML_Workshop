```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay


# --- Step 1: Feature Engineering and Clean Dataset ---
#To make a copy of  original dataset df to work safely
df_b = df.copy()

#create same features
df_b['past_performance'] = (df_b['G1'] + df_b['G2']) / 2
df_b['study_efficiency'] = df_b['past_performance'] / (df_b['studytime'] + 1)
df_b['parent_edu'] = (df_b['Medu'] + df_b['Fedu']) / 2

#I picked only relevant variables
df_b = df_b[['age', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout',
             'Dalc', 'Walc', 'health', 'absences', 'past_performance', 'study_efficiency',
             'parent_edu', 'school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob',
             'Fjob', 'reason', 'guardian', 'schoolsup', 'famsup', 'paid', 'activities',
             'nursery', 'higher', 'internet', 'romantic', 'pass']]


#seperate features and target
X_b = df_b.drop(columns='pass')
y_b = df_b['pass']
```

```python
# --- Step 2: Preprocessing and Pipelines ---
num_cols_b = ['age', 'traveltime', 'studytime', 'failures', 'famrel', 'freetime', 'goout',
              'Dalc', 'Walc', 'health', 'absences', 'past_performance', 'study_efficiency', 'parent_edu']
cat_cols_b = ['school', 'sex', 'address', 'famsize', 'Pstatus', 'Mjob', 'Fjob', 'reason',
              'guardian', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher',
              'internet', 'romantic']

#processing pipeline, standardize numeric columns and one-hot encode categorical columns and we are dropping first ones to avoid multicollinearity
preprocessor_b = ColumnTransformer([
    ('num', StandardScaler(), num_cols_b),
    ('cat', OneHotEncoder(drop='first'), cat_cols_b)
])

# Pipelines for each model with balanced class weight where applicable
#used preprocessor_b for scaled models and use raw data for tree models(dt and rf since they handle categories and scales)
pipelines = {
    'Logistic Regression': Pipeline([('preprocessor', preprocessor_b), ('clf', LogisticRegression(max_iter=1000, class_weight='balanced'))]),
    'KNN': Pipeline([('preprocessor', preprocessor_b), ('clf', KNeighborsClassifier())]),
    'SVM': Pipeline([('preprocessor', preprocessor_b), ('clf', SVC(class_weight='balanced', probability=True))]),
    'Decision Tree': Pipeline([('clf', DecisionTreeClassifier(class_weight='balanced', random_state=42))]),
    'Random Forest': Pipeline([('clf', RandomForestClassifier(class_weight='balanced', random_state=42))])
}
```

```python
# --- Step 3: Hyperparameter Grids ---
param_grids = {
    'Logistic Regression': {'clf__C': [0.01, 0.1, 1, 10, 100]},
    'KNN': {'clf__n_neighbors': [3, 5, 7, 9], 'clf__weights': ['uniform', 'distance']},
    'SVM': {'clf__C': [0.1, 1, 10], 'clf__gamma': ['scale', 'auto'], 'clf__kernel': ['rbf', 'linear']},
    'Decision Tree': {'clf__max_depth': [None, 5, 10, 20], 'clf__min_samples_split': [2, 5, 10]},
    'Random Forest': {'clf__n_estimators': [100, 200], 'clf__max_depth': [None, 5, 10], 'clf__min_samples_split': [2, 5, 10]}
}
```

```python
# Step 4 plit properly
## X_train_b contains raw features for scaled models (preprocessing inside pipeline), while X_train_A is already one-hot encoded and numeric for tree models that do not require scaling.

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_b, y_b, test_size=0.2, random_state=42, stratify=y_b)
X_train_A, X_test_A, y_train_A, y_test_A = train_test_split(X_A, y_A, test_size=0.2, random_state=42, stratify=y_A)

```

```python
# --- Step 5: GridSearchCV (Only on training data) ---
#stratifies k fold keeps balance as is
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
best_models = {}

#this for loop perform GridSearchCV on each model but only for training set and then
#selects best model and keep it in best_models
#it uses f1 as the storing metrics
for name in pipelines:
    print(f"\nTraining {name}...")
    grid = GridSearchCV(pipelines[name], param_grids[name], cv=cv, scoring='f1', n_jobs=-1)
    if name in ['Decision Tree', 'Random Forest']:
        grid.fit(X_train_A, y_train_A)  # Tree models on already encoded data
    else:
        grid.fit(X_train_b, y_train_b)  # Scaled models on raw df_b
    best_models[name] = grid.best_estimator_
    print(f"Best params: {grid.best_params_}, Best F1 (CV): {grid.best_score_:.3f}")
```

```python
# Evaluation on test set with correct datasets
#here we are using the tuned modes to make predictions on the test set
#again X_test_b for scaled models X_test_A for tree models
preds = {
    'Logistic Regression': best_models['Logistic Regression'].predict(X_test_b),
    'KNN': best_models['KNN'].predict(X_test_b),
    'SVM': best_models['SVM'].predict(X_test_b),
    'Decision Tree': best_models['Decision Tree'].predict(X_test_A),
    'Random Forest': best_models['Random Forest'].predict(X_test_A)
}
```

```python
# Plot confusion matrices for all models
#to see how many fails and passes were correctly predicted by each model
fig, axes = plt.subplots(1, 5, figsize=(20, 4))
for ax, (name, y_pred) in zip(axes, preds.items()):
    cm = confusion_matrix(y_test_b if name in ['Logistic Regression', 'KNN', 'SVM'] else y_test_A,
                          y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Fail', 'Pass'])
    disp.plot(ax=ax, colorbar=False)
    ax.set_title(name)
plt.tight_layout()
plt.show()
```

```python
# Classification reports and summary table

#create initial empty dictionary to store classification reports of each model, key is model name and valu is the classification report
reports = {}

#iterates over the dictionary preds
for name, y_pred in preds.items():
    y_true = y_test_b if name in ['Logistic Regression', 'KNN', 'SVM'] else y_test_A #if scaled model use y_text_b if tree model use y_test_A
    report = classification_report(y_true, y_pred, target_names=['Fail', 'Pass'], output_dict=True)  #generates classification report and labes fail as 0 and pass as 1
    reports[name] = report

# Create summary table
summary = []
for name, report in reports.items():
    summary.append({
        'Model': name,
        'F1 Fail': round(report['Fail']['f1-score'], 3),
        'Recall Fail': round(report['Fail']['recall'], 3),
        'Accuracy': round(report['accuracy'], 3)
    })

summary_df = pd.DataFrame(summary).sort_values('F1 Fail', ascending=False).reset_index(drop=True)
print(summary_df)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

plt.figure(figsize=(10, 8))

#loop over all the best tuned models that we obtained from GridSearchCV
for name, model in best_models.items():
    # Use correct test set per model
    X_test = X_test_b if name in ['Logistic Regression', 'KNN', 'SVM'] else X_test_A
    y_test = y_test_b if name in ['Logistic Regression', 'KNN', 'SVM'] else y_test_A

    # Check if model has predict_proba or decision_function
    if hasattr(model, "predict_proba"):
        y_score = model.predict_proba(X_test)[:, 1]  # probability for class 1 ('Pass')
    elif hasattr(model, "decision_function"):
        y_score = model.decision_function(X_test)
    else:
        continue  # Skip if model cannot produce probabilities

    # Calculate ROC curve and AUC
    #calculate false positive rate, true positive rate, and threshold
    fpr, tpr, thresholds = roc_curve(y_test, y_score) #input is true labels and predicted probabilities are y_score
    auc_score = roc_auc_score(y_test, y_score)  #calculates area under the curve

    # Plot the ROC curve and label the model names
    plt.plot(fpr, tpr, label=f'{name} (AUC = {auc_score:.3f})')

# Plot settings
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')
plt.title('ROC-AUC Curve Comparison')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid(True)
plt.show()
```
